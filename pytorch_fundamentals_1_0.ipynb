### tensor aggregation

```python
tensor_agg = torch.rand(3,3)
print(tensor_agg)
print(f"tensor dtype: {tensor_agg.dtype}")
print(f"min value : {torch.min(tensor_agg)}")
print(f"max value : {torch.max(tensor_agg)}")
print(f"mean value : {torch.mean(tensor_agg)}")
print(f"sum : {torch.sum(tensor_agg)}")
print(f"positional min : {torch.argmin(tensor_agg)}")
print(f"positional max : {torch.argmax(tensor_agg)}")
```

### tensor reshaping, stacking, squeezing, unsqueezing, permute

```python
tensor_X =torch.rand(1,10)
print(tensor_X)
print(f"tensor dtype : {tensor_X.dtype}")
print(f"\nshape : {tensor_X.shape}")
print(f"\nreshape : {tensor_X.reshape(10,1)}")
print(f"\nstacks : {torch.stack([tensor_X, tensor_X, tensor_X], dim=0)}")
print(f"\nsequeeze : {torch.squeeze(tensor_X)} \nsqueeze size :  {torch.squeeze(tensor_X).shape}")
print(f"\nunsqueeze : {torch.unsqueeze(tensor_X, dim=0)}")
print(f"\npermute : {torch.permute(tensor_X, (-2,1))}")
```

### pytorch Reproducibility - reduce randomness

```python
random_seed = 42
torch.manual_seed(random_seed)
tensorA = torch.rand(3,3)

torch.manual_seed(random_seed)
tensorB = torch.rand(3,3)

print(tensorA)
print(tensorB)
print(tensorA == tensorB)
```
